```python
import os
import torch
import pandas as pd

from glob import glob

from sklearn.model_selection import train_test_split

```

```python
total_data = []
for i in glob('../dataset_v2/*'):
    total_data.extend(glob(i + '/*'))
```

```python
df = pd.DataFrame({'data' : total_data, 'label': list(map(lambda x : x.split('/')[2], total_data))})
```

```python
df['label'].value_counts()
```

```python
train, valid = train_test_split(df, test_size=0.3)
test, valid = train_test_split(valid, test_size = 2/3)

print(f'{len(train), len(valid), len(test)}')
```
